```python
# Read output.csv and plot gantt chart for the data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/workspaces/DSA/activity-selection-algorithm/data/output.csv')



total_profit = df['Profit']
start_times = df['Start Time']
end_times = df['Finish Time']
acticity_names = df['Activity Name']

# convert to arrays
total_profit = np.array(total_profit)
start_times = np.array(start_times)
end_times = np.array(end_times)
acticity_names = np.array(acticity_names)


# Your code to read the CSV and convert columns to arrays

# Create a Gantt chart
fig, ax = plt.subplots(figsize=(10, 6))

# Loop through the activities and plot them on the Gantt chart
for i in range(len(acticity_names)):
    ax.barh(acticity_names[i], width=end_times[i] - start_times[i], left=start_times[i], label=acticity_names[i])

# Set labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Activities')
ax.set_title('Gantt Chart')

# Add a legend
ax.legend()

# Show the Gantt chart
plt.show()



```
