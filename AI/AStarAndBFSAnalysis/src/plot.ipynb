```python
import matplotlib.pyplot as plt

def read_results():
    results = []
    with open('results.txt', 'r') as f:
        next(f)  
        for line in f:
            results.append([int(x) for x in line.split() if x.isdigit()])
    return results

def plot_results(results):
    x = [x[0] for x in results]  
    y1 = [x[1] for x in results] 
    y2 = [x[2] for x in results]  
    y3 = [x[3] for x in results]  
    plt.plot(x, y1, label='A* (Manhattan Distance)')
    plt.plot(x, y2, label='A* (Misplaced Tiles)')
    plt.plot(x, y3, label='BFS')
    plt.xlabel('Inversion Count')
    plt.ylabel('States Generated')
    plt.title('States Generated vs Inversion Count')
    plt.legend()
    plt.show()

if __name__ == '__main__':
    results = read_results()
    plot_results(results)
```
