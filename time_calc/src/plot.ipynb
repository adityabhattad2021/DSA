```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/workspaces/DSA/time_calc/data/observations.csv')
```

```python
x = df['length']
y1 = df['time for divide and conquer']
y2 = df['time of brute force']

# Create a line graph
plt.figure(figsize=(10, 6))

# Plotting two line plots
plt.plot(x, y1, marker='o', linestyle='-', label='Divide and Conquer')
plt.plot(x, y2, marker='o', linestyle='-', label='Brute Force')

# Customize the plot
plt.title('Time vs. Length')
plt.xlabel('Length')
plt.ylabel('Time (milliseconds)')
plt.legend() # Automatic legend from labels in plt.plot()
plt.grid(True)

# Show the plot
plt.show()
```
